# DATA VISUALIZATION ON CLASS DATASET

Step1 = Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Presentation-TASK - Sheet1.csv")
df.head()
```

```python
df.tail()
```

```python
# lets see how many null values are there
df.isnull().sum()
```

```python
# lets see the data types of each column
print(df.dtypes)
```

```python
# in above we can see that technical is in object so conver into numeric
df["Technical knowledge"] = pd.to_numeric(df["Technical knowledge"], errors="coerce")
```

```python
# now we have to remove "marks column" bcz we cant remove null values
df = df.drop("Marks", axis=1)
```

```python
df.head(3)
```

```python
# now remove null values
df = df.dropna()
df.isnull().sum()
```

```python
#Now add lets create column called Total Marks
df["Total Marks"] = df["Technical knowledge"] + df["Communication"] + df["Attrire"]
```

```python
df.info()
```

```python
# check how many null values are there
df.isnull().sum()
```

```python
df.head(5)
```

# Data Visualization


Top 5 students based on Total Marks

```python
# Top 5 Total Marks
#in the below code we sort itb in a decending order and take first 5 students so it gives only 5 names
top_total = df.sort_values("Total Marks", ascending=False).head(5)
plt.bar(top_total["Participant Name"], top_total["Total Marks"])

plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.title("Top 5 Students Based on Total Marks")
plt.xticks(rotation=45)#Rotates student names because long names overlap.

plt.show()
```

Top 5 students in Technical

```python
top_tech = df.sort_values("Technical knowledge", ascending=False).head(5)

plt.bar(top_tech["Participant Name"],top_tech["Technical knowledge"])
plt.xlabel("Student Name")
plt.ylabel("Technical Marks")
plt.title("Top 5 Technical Students")

plt.xticks(rotation=45)

plt.show()
```

Top 5 students in Communication


```python
top_comm = df.sort_values("Communication", ascending=False).head(5)
# there are may people in top 5
plt.bar(top_comm["Participant Name"],top_comm["Communication"])
plt.xlabel("Student Name")
plt.ylabel("Communication Marks")
plt.title("Top 5 Communication Students")
plt.xticks(rotation=45)
plt.show()
```

Top 5 students in Attire

```python
top_attire = df.sort_values("Attrire", ascending=False).head(5)

plt.bar(top_attire["Participant Name"],top_attire["Attrire"])
plt.xlabel("Student Name")
plt.ylabel("Attire Marks")
plt.title("Top 5 Attire Students")
plt.xticks(rotation=45)
plt.show()
```
