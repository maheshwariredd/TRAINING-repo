Which YouTube channels have the highest number of subscribers?  

```python
# for this we should arrenge the data in decending order
top_subscribers = df.sort_values(by="subscribers", ascending=False)
```

```python
# now  by arrenging in decending oredr we can find the top 10 subscribers
top_subscribers.head(10)
```

```python
top_subscribers[["Youtuber", "subscribers"]].head(10) # if we want a specific rows we use it
```

```python
# barplot fro it
plt.figure(figsize=(4,2))

sns.barplot(
    data=top_subscribers.head(10),
    x="subscribers",
    y="Youtuber"
)

plt.title("Top 10 YouTube Channels by Subscribers")
plt.xlabel("Subscribers")
plt.ylabel("YouTube Channel")

plt.show()
```

Which channels have the highest video views?



```python
top_views = df.sort_values(by="video views", ascending=False)
```

```python
top_views[["Youtuber", "video views"]].head(10)
```

```python
plt.figure(figsize=(4,2))

sns.barplot(
    data=top_views.head(10),
    x="video views",
    y="Youtuber"
)

plt.title("Top 10 YouTube Channels by Video Views")
plt.xlabel("Video Views")
plt.ylabel("YouTube Channel")

plt.show()
```

Which countries have the highest number of YouTube creators?

```python
country_count = df["Country"].value_counts()
```

```python
country_count.head(10)
```

```python
plt.figure(figsize=(5,4))

sns.barplot(
    x=country_count.head(10).values,
    y=country_count.head(10).index
)

plt.title("Top 10 Countries with Highest Number of YouTube Creators")
plt.xlabel("Number of Creators")
plt.ylabel("Country")

plt.show()
```
