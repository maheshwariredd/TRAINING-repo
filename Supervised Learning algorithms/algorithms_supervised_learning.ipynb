**SVM Algorithm**


```python
import pandas as pd
from sklearn.svm import SVC

# Dataset
df = pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Pass":[0,0,0,1,1,1]
  })
#Independent variable
X = df[["Hours"]]
#dependent variable
y = df["Pass"]
#create a model
model = SVC(kernel="linear")
#train model
model.fit(X, y)
# these are the prediction process
prediction = model.predict([[4.5]])
print("Prediction:", prediction[0])
print("Slope:", model.coef_[0][0])

print("Intercept:", model.intercept_[0])
```

**KNN**

```python
import pandas as pd
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
# here 1 = Male
#      0 = Female
# Dataset
df = pd.DataFrame({
    "Height":[150,155,160,165,170,175],
    "Gender":[0,0,0,1,1,1]
    })
# Independent variable
X = df[["Height"]]
# dependent variable
y = df["Gender"]

# Create a model this model checks the 3 nearest data points before making a prediction.
model = KNeighborsClassifier(n_neighbors=3)

# Training the model
model.fit(X, y)

#prediction process
prediction = model.predict([[168]])

print("Predicted Gender:", prediction[0])
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# create a data set
df = pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Pass" : [0,0,0,1,1,1]
})

#Independent variable
X = df[["Hours"]]

#dependent variable
y = df["Pass"]

#create a model here (n_estimators=10) means the model builds 10 decision trees.
# random_state=42 ensures you get the same result every time you run the code.
model =  RandomForestClassifier(n_estimators=10, random_state=42)

#Training the model
model.fit(X,y)

#Pridiction
Prediction = model.predict([[5]])

print("Prediction:", prediction[0])
```

**Polynomial Regression **

y=ax2 + bx + c
Polynomial Regression fits a curve, giving more accurate predictions.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Dataset
df = pd.DataFrame({
    "Hours":[1,2,3,4,5],
    "Score":[50,55,65,80,100]
        })
# Independent Variable
X = df[["Hours"]]

# Dependent Variable
y = df["Score"]

# Convert Linear Data into Polynomial Data
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Create Model
model = LinearRegression()

# Train Model
model.fit(X_poly, y)

# Prediction
prediction = model.predict(poly.transform([[6]]))

print("Predicted Score:", prediction[0])

print("Slope:", model.coef_)

print("Intercept:", model.intercept_)
```
