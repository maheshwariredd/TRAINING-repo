# **Introduction to Reinforcement Learning (RL)**

Reinforcement Learning is a learning technique in which an agent learns the best actions by receiving rewards and penalties from the environm

Instead of learning from labeled data like supervised learning, the agent learns through trial and error.

Whenever the agent performs an action:

It receives a reward if the action is good.

It receives a penalty if the action is bad.

The main goal of the agent is to maximize the total reward over time.

**Types of Reinforcement Learning**

Positive Reinforcement
Reward is given.

Example:

Correct Answer → +10



Negative Reinforcement
Penalty is removed.

Example:

Seat belt alarm stops after fastening the seat belt.

**Applications** :

Self-driving cars

Robotics

Chess

Video Games

Stock Trading

Traffic Signal Control

Recommendation Systems



```python
# code
#random module to choose actions randomly.
import random

#it creats 2 Possible actions
actions = ["Left", "Right"]

# Stores the cumulative reward.
total_reward = 0

print("Reinforcement Learning Simulation\n")

# Run for 10 steps
for step in range(1, 11):

     #Randomly selects either Left or Right.
    action = random.choice(actions)

if action == "Right":
        reward = 10
else:
        reward = -5

total_reward += reward
print("Step:", step)
print("Action:", action)
print("Reward:", reward)
print()
print("Total Reward:", total_reward)
```

# **Markov Decision Processes (MDP)**

A Markov Decision Process (MDP) is a mathematical model used in Reinforcement Learning (RL) to describe how an agent makes decisions in an environment.

An MDP helps an agent decide which action to take in each state so that it can maximize its total reward over time.

The Markov Property states that the next state depends only on the current state and the current action, not on the previous states.

**Markov Property**

The most important property of an MDP is the Markov Property.

The future depends only on the current state and current action.

It does not depend on earlier states.

Example:

Current Position → Move Right → Next Position

The agent does not need to remember every previous position.

**Real-Life Example**

Imagine you are using Google Maps to travel from your home to your college.

State: Your current location.

Action: Move left, move right, or go straight.

Reward: You move closer to your destination.

Penalty: You take a wrong turn or get stuck in traffic.

Goal: Reach the destination in the shortest time.

At every step, you decide based only on your current location, not on the entire path you have already traveled. This follows the Markov Property.

```python
#code
# Define the states
states = ["Room A", "Room B", "Goal"]

# Define rewards for each state
rewards = {
    "Room A": 0,
    "Room B": 5,
    "Goal": 10
            }

# Define transitions
transitions = {
   "Room A": "Room B",
   "Room B": "Goal"
 }

 # Start from Room A
current_state = "Room A"
print("Markov Decision Process Simulation\n")

# Continue until Goal is reached
while current_state != "Goal":

    print("Current State:", current_state)

    next_state = transitions[current_state]

    print("Move To:", next_state)

    print("Reward Received:", rewards[next_state])
    print()
    current_state = next_state

    print("Current State:", current_state)
    print("Goal Reached!")
    print("Final Reward:", rewards[current_state])
```

# Q-Learning Algorithm
Q-Learning is a model-free Reinforcement Learning algorithm that helps an agent learn the best action to take in each state without knowing how the environment works.

The agent learns by interacting with the environment and storing its experience in a Q-Table. Over time, it updates the Q-values and eventually learns the best path to maximize rewards.



**Q-Learning Formula**

The Q-value is updated using the following formula:

Q(s,a) = Q(s,a) + α [ R + γ × max(Q(s',a')) − Q(s,a) ]

Where:

Q(s,a) = Current Q-value

α (Alpha) = Learning rate

R = Reward

γ (Gamma) = Discount factor

max(Q(s',a')) = Maximum future reward

s' = Next state

**Real-Life Example**

Imagine you are learning the fastest route from your home to college.

On the first day, you may take a wrong road.

On the next day, you try another route.

Every day, you remember which roads save time and which roads cause delays.

After several days, you always choose the fastest route.

```python
#code
import numpy as np

# Create a Q-Table (3 states × 2 actions)
Q = np.zeros((3, 2))

# Learning parameters
alpha = 0.5      # Learning rate
gamma = 0.9      # Discount factor

# Current state and action
state = 0
action = 1

# Reward received
reward = 10

# Next state
next_state = 1

# Update Q-value
Q[state, action] = Q[state, action] + alpha * (
    reward + gamma * np.max(Q[next_state]) - Q[state, action]
    )

print("Updated Q-Table:")
print(Q)
```

# Deep Q-Networks (DQN)

A Deep Q-Network (DQN) is an advanced Reinforcement Learning algorithm that combines Q-Learning with Deep Learning.

In traditional Q-Learning, the agent stores values in a Q-Table. However, when the environment has a very large number of states (such as images or video games), creating a Q-Table becomes impossible.

A DQN solves this problem by using a Neural Network instead of a Q-Table. The neural network predicts the Q-values for each action based on the current state.

**Real-Life Example**

Imagine a self-driving car learning to drive.

The car receives images from its camera.

It decides whether to turn left, turn right, speed up, or slow down.

Safe driving gives positive rewards.

Collisions give negative rewards.

Over time, the neural network learns the best driving strategy.

**Applications**

Self-driving cars

Robotics

Healthcare

Video games (Atari)

Drone navigation

Smart traffic control

Warehouse automation

**Why Do We Need DQN?**

Traditional Q-Learning works well for small problems.

Example:

5 Rooms

2 Actions

A Q-Table is enough.

But consider:

Thousands of game screens

Millions of images

Self-driving car cameras

A Q-Table becomes too large.

Therefore, we use a Deep Neural Network to estimate Q-values.

```python
#code
# TensorFlow library used for building deep learning models.
import tensorflow as tf

# Sequential model, which allows layers to be added one after another.
from tensorflow.keras.models import Sequential

#Dense (fully connected) layer used to build the neural network.
from tensorflow.keras.layers import Dense
import numpy as np

# Create DQN model
model = Sequential([
    Dense(24, input_shape=(4,), activation='relu'),
    Dense(24, activation='relu'),
    Dense(2, activation='linear')
   ])

# Compile model
model.compile(optimizer='adam', loss='mse')

# Sample state
state = np.array([[1, 0, 1, 0]])

# Predict Q-values
q_values = model.predict(state, verbose=0)

print("Predicted Q-values:")
print(q_values)
```
