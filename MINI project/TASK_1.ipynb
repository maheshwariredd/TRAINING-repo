Create a Student Report Card Program using:

1.List → store student names # here names can be same

2.Set → store subjects # here no subject should be repeted

3.Tuple → store student information (name, age)

4.Dictionary → store subject-wise marks

```python
Student_name = [] # list
Student_sub = set() # empty set so we use () set with values we take {}
Student_info = () # tuple
Student_mark = {} # we use this symbole for empty dictionary

print("===== Enter Student Details =====")
name = input("Enter student name : ")
age = input ("Enter student age : ")
num_sub = int(input("Enter no.subjects : "))

Student_name.append(name)
Student_info = (name, age)

#For each subject enter name and marks
for i in range(num_sub):
  print(f"\nSubject {i+1} :")
  subject = input("  Enter Subject Name : ")
  marks   = float(input("  Enter Marks        : "))

  Student_sub.add(subject)
  Student_mark[subject] = marks

# Calculate Total and Average
total   = sum(Student_mark.values()) # it is out side the loop
average = total / num_sub

print("=" * 35)
print("     STUDENT REPORT CARD")
print("=" * 35) #Print the = symbol 35 times in a single line

print("Name     :", Student_info[0])

print("Age      :", Student_info[1])
print("-" * 35)

print("Subjects :", Student_sub)
print("-" * 35)

print("Marks :")
for subject, marks in Student_mark.items():
  print(" ", subject, ":", marks)
  print("-" * 35)

    # Print Total and Average
  print("Total Marks   :", total)
  print("Average Marks :", average)
  print("=" * 35)
```
